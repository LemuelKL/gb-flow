```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    "InterconnectorFlows-2016-01-01T00_00_00.000Z-2024-11-23T20_30_00.000Z.csv"
)
```

```python
df
```

```python
df = df.drop(columns=["Dataset", "PublishTime", "StartTime"])
df = df.rename(columns={"Generation": "Flow"})
df["FuelType"] = df["FuelType"].str.replace("INT", "")
df = df.rename(columns={"FuelType": "Interconnector"})
```

```python
df.describe()
```

```python
print(df["Interconnector"].unique())
```

```python
df
```

```python
df["SettlementDate"] = pd.to_datetime(df["SettlementDate"])
df["Year"] = df["SettlementDate"].dt.year
df["Month"] = df["SettlementDate"].dt.month
df["Day"] = df["SettlementDate"].dt.day
df = df.drop(columns=["SettlementDate"])
df
```

```python
df_by_interconnector = df.pivot_table(
    index=["Year", "Month", "Day", "SettlementPeriod"], columns="Interconnector", values="Flow"
).reset_index()
df_by_interconnector
```

```python
df_by_interconnector.to_csv("interconnector_flows.csv", index=False)
```
